### Project : City Streetlight Energy Consumption Analysis

### Objectives
- Analyze hourly streetlight energy consumption
- Identify peak usage hours
- Understand energy usage patterns for optimization

### Tools and Technologies Used
- Python
- Pandas
- Matplotlib
- Jupyter Notebook



```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("intelligent_streetlight_dataset.csv")
data.head()
```

```python
# Converting timestamp column to datetime and extract hour
data['Timestamp'] = pd.to_datetime(data['Timestamp'])
data['Hour'] = data['Timestamp'].dt.hour
```

```python
# Summary statistics of energy consumption
data['Energy Consumption (kWh)'].describe()
```

```python
# Hour-wise average energy consumption
peak_usage = data.groupby('Hour')['Energy Consumption (kWh)'].mean()
peak_usage
```

```python
plt.figure()
peak_usage.plot(kind='line')
plt.xlabel('Hour of Day')
plt.ylabel('Average Energy Consumption')
plt.title('Peak Energy Consumption Hours')
plt.show()
```

```python
# Top 3 peak energy consumption hours
top_3_hours = peak_usage.sort_values(ascending=False).head(3)
top_3_hours
```

### Conclusion
The analysis successfully identified peak energy consumption hours in the city streetlight system. The results indicate that most energy is consumed during night hours, emphasizing the importance of smart lighting solutions. By applying data analytics techniques, cities can optimize energy usage, reduce power wastage, and promote sustainable urban development.
